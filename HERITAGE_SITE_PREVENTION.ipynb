```python
# Heritage Site Preservation Project
# Name: Adhithi Madhav S
# Course: Mathematics 1st DC

# Function to calculate decay rate
def calculate_decay(age, erosion_factor, human_impact):
    decay_rate = (age * erosion_factor) + human_impact
    return decay_rate

# Function to determine preservation status
def preservation_status(decay_rate):
    if decay_rate < 50:
        return "Good Condition"
    elif decay_rate < 100:
        return "Moderate Condition"
    else:
        return "Critical Condition"

# Function to suggest best visiting time
def best_time_to_visit(status):
    if status == "Good Condition":
        return "Any time of the year"
    elif status == "Moderate Condition":
        return "Visit during non-monsoon season"
    else:
        return "Restricted visit – Preservation ongoing"

# Input data (replaced input() with hardcoded values for demonstration)
monument_name = "Eiffel Tower"
age = 134  # Example age in years
erosion_factor = 2.5  # Example natural erosion factor
human_impact = 3.8  # Example human impact factor

# Processing
decay = calculate_decay(age, erosion_factor, human_impact)
status = preservation_status(decay)
visit_time = best_time_to_visit(status)

# Output
print("\n--- Heritage Site Preservation Report ---")
print("Monument:", monument_name)
print("Decay Rate:", decay)
print("Preservation Status:", status)
print("Suggested Visit Time:", visit_time)
```
